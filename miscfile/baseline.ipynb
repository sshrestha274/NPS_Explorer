```python

# Create DataFrame
columns = ["effect", "category", "polarity", "example_phrases"]
df = pd.DataFrame(nps_effects_data, columns=columns)

```

```python
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

# Find similar or duplicate effects
similar_pairs = []
checked = set()

for i in range(len(df)):
    effect_i = df.iloc[i]['effect'].lower()
    for j in range(i + 1, len(df)):
        effect_j = df.iloc[j]['effect'].lower()
        if (effect_i, effect_j) in checked or (effect_j, effect_i) in checked:
            continue
        if similar(effect_i, effect_j) > 0.8:
            similar_pairs.append((df.iloc[i]['effect'], df.iloc[j]['effect']))
            checked.add((effect_i, effect_j))

similar_pairs_df = pd.DataFrame(similar_pairs, columns=["Effect 1", "Effect 2"])
similar_pairs_df
```
